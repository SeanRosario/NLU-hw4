```python
def plot_curves(fname):
    with open(fname) as f:
        content = f.readlines()

    train_perp=[]
    b0=[]
    b1=[]
    b2=[]
    b3=[]
    for line in content:
        if line.split()[0]=='global':
            train_perp.append(float(line.split()[-1]))

    for line in content:
        x= line.split()
        if x[0]=='eval:':
            if x[2]=='0':
                b0.append(x[4])
            if x[2]=='1':
                b1.append(x[4])
            if x[2]=='2':
                b2.append(x[4])
            if x[2]=='3':
                b3.append(x[4])


    t=np.arange(len(b1))
    #plt.plot(t, train_perp,  t, b0,  t, b1,t, b2,t, b3)
    plt.plot(t,train_perp,label="train perplexity",alpha=0.5,lw=2)
    plt.plot(t,b0,label="eval on bucket 0",alpha=0.5)
    plt.plot(t,b1,label="eval on bucket 1",alpha=0.5)
    plt.plot(t,b2,label="eval on bucket 2",alpha=0.5)
    plt.plot(t,b3,label="eval on bucket 3",alpha=0.5)
    plt.legend()
    plt.savefig(fname+".png")
```

```python
plot_curves("without_attention.txt")
```

```python
plot_curves("with_attention.txt")
```
